```python
def addCountDict(key, my_dict):
    try:
        my_dict[key] += 1
    except:
        my_dict[key] = 1
```

```python
# from PIL import Image
# def rgb_of_pixel(img_path, x, y):
im = Image.open(flag1Name).convert('RGB')
r, g, b = im.getpixel((0, 0))
imArr = np.asarray(im)
print(imArr.shape)
print(imArr[0][0])
print(imArr[0][1])
print(imArr[1][1])
```

```python
(M,N,z) = imArr.shape
flagA = {}
for i in range(M):
    for j in range(N):
        pixel = imArr[i][j]
        t = (pixel[0], pixel[1], pixel[2])
        addCountDict(t,flagA)
```
